```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_num_cuts = [5, 20, 30, 40]
all_num_requests = [100]

for num_cuts in all_num_cuts:
    for num_requests in all_num_requests:
        file_path = f'loss_{num_cuts}_{num_requests}.npy'
        losses = np.load(file_path, allow_pickle=True)

        plt.figure()
        for loss in losses:
            plt.plot(loss)

        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title(f'{num_cuts} Cuts, Best Loss: {np.min(np.concatenate(losses)):.2f}')
        plt.legend([f'Min Loss: {np.min(losses[i]):.2f}' for i in range(len(losses))])
        plt.savefig(f'loss_{num_cuts}_{num_requests}.png')
        plt.show()
```
