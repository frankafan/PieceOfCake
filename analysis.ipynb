```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_num_cuts = [6, 7, 8, 9, 10, 11, 12]
all_num_requests = [10]

for num_cuts in all_num_cuts:
    for num_requests in all_num_requests:
        file_path = f'loss_{num_cuts}_{num_requests}.npy'
        losses = np.load(file_path, allow_pickle=True)

        plt.figure()
        for loss in losses:
            plt.plot(loss)

        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title(f'{num_cuts} Cuts, Best Loss: {np.min(np.concatenate(losses)):.2f}')
        plt.legend([f'Min Loss: {np.min(losses[i]):.2f}' for i in range(len(losses))])
        plt.savefig(f'loss_{num_cuts}_{num_requests}.png')
        plt.show()
```

```python
for num_requests in all_num_requests:
    plt.figure()

    min_losses = []
    for num_cuts in all_num_cuts:
        file_path = f'loss_{num_cuts}_{num_requests}.npy'
        losses = np.load(file_path, allow_pickle=True)
        min_loss = np.min(np.concatenate(losses))
        min_losses.append(min_loss)

    plt.plot(all_num_cuts, min_losses)
    plt.xlabel('Number of Cuts')
    plt.ylabel('Best Loss')
    plt.title(f'{num_requests} Requests')
plt.show()
```
